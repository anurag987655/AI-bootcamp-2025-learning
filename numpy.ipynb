```python
#exercise 2 :implement an image filter using numpy array manipulation:


import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img = mpimg.imread("luffy.jpeg")

height =img.shape[0]
width=img.shape[1]

img_inv = 255 - img
img_rot = np.rot90(img)
flip_h=np.flip(img,axis=1)
flip_v=np.flip(img,axis=0)
diagonal_flip=np.flip(np.transpose(img, (1,0,2)), axis=1)


plt.subplot(3,2,1); plt.title("Original"); plt.imshow(img); plt.axis('off')
plt.subplot(3,2,2); plt.title("Diagonal Flip"); plt.imshow(diagonal_flip); plt.axis('off')
plt.subplot(3,2,3); plt.title("Inverted"); plt.imshow(img_inv); plt.axis('off')
plt.subplot(3,2,4); plt.title("Rotated 90Â°"); plt.imshow(img_rot); plt.axis('off')
plt.subplot(3,2,5); plt.title("Horizontal flip"); plt.imshow(flip_h); plt.axis('off')
plt.subplot(3,2,6); plt.title("Vertical flip"); plt.imshow(flip_v); plt.axis('off')
plt.show()
```

```python
#implement 2d rotation of image using numpy

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img=mpimg.imread("luffy.jpeg")

height=img.shape[0]
width=img.shape[1]

#numpy has builtin function that allows us to rotate image in multiple of 90 degree ccw

rot_90  = np.rot90(img, k=1)
rot_180 = np.rot90(img, k=2)
rot_270 = np.rot90(img, k=3)

plt.subplot(1,4,1);plt.title("Original");plt.imshow(img);plt.axis('off')
plt.subplot(1,4,2);plt.title("Rotated 90");plt.imshow(rot_90);plt.axis('off')
plt.subplot(1,4,3);plt.title("Rotated 180");plt.imshow(rot_180);plt.axis('off')
plt.subplot(1,4,4);plt.title("Rotated 270");plt.imshow(rot_270);plt.axis('off')
plt.show()


```

```python
# solve a 3*3 system to verify A@x=b

import numpy as np
A=np.array([[1,2,-2],[2,1,-5],[1,-4,1]])
b=np.array([-15,-21,18])

x=np.linalg.solve(A,b)

print("The solution is :",x)

#to verify that A x =b
b= A @ x
print("The value of b is :",b)

```

```python
#horizentally stack 2 3*2 matrix into 3*4 matrix

import numpy as np
a=np.ones((3,2))
print("a is ")
print(a)
b=np.zeros((3,2))
print("b is ")
print(b)

c=np.hstack((a,b))
print("c is 3*4 matrix:")
print(c)
```

```python
# compute column wise mean and subtract it from matrix(centering)

import numpy as np

X=np.arange(1,10).reshape(3,3)
print("Original Matrix:")
print(X)

mean=X.mean(axis=0)
result = X - mean

print("Mean: ")
print(mean)
print("Resulting Matrix:")
print(result)

```
